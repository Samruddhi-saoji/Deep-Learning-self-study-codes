```python
import numpy as np
import matplotlib.pyplot as plt
```

# Glorot/Xavier

## Weight distributions

```python
# returns True if x is in range [a, b]
def in_range(x, a, b):
   if x<=b and x>=a:
     return True
   return False

x = 10 # std dev
n = 500 #num of points in the distribution
x_range = [-3.5*x, 3.5*x] #range of x coordinates
x_vals = np.linspace( x_range[0], x_range[1], n)

# Normal Distribution
mu = 0  # mean
sigma = x  # standard deviation
#n_values = np.random.normal(mu, sigma, 1000) #values in the distribution
# probability distribution function
normal_pdf = (1 / (np.sqrt(2 * np.pi) * sigma)) * np.exp(-0.5 * ((x_vals - mu) / sigma) ** 2)


# Uniform Distribution
# range = [min, max]
min = -x * np.sqrt(3)
max = x * np.sqrt(3)
range = 2*max #max-min
#u_values = np.random.uniform(min, max, 1000)
uniform_pdf = [1/range if in_range(val, min, max) else 0 for val in x_vals]
```

```python
# separate graphs
# Plotting Normal Distribution PDF
plt.figure(figsize=(8, 4))
plt.plot(x_vals, normal_pdf, label='Normal Distribution', color='blue')
plt.title('Normal Distribution with $\mu=0$ and $\sigma=x$')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.grid(True)
plt.show()

# Plotting Uniform Distribution PDF
plt.figure(figsize=(8, 4))
plt.plot(x_vals, uniform_pdf, label='Uniform Distribution', color='orange')
plt.title('Uniform Distribution in range $[-x\sqrt{3}, x\sqrt{3}]$')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.grid(True)
plt.show()
```

```python
# single graph
plt.figure(figsize=(8, 4))

# Plot Normal Distribution PDF
plt.plot(x_vals, normal_pdf, label='Normal Distribution')

# Plot Uniform Distribution PDF
plt.plot(x_vals, uniform_pdf, label='Uniform Distribution')

plt.title('Normal and Uniform Distributions')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```
